```python
fpath = ps.examples.get_path('Chicago77.shp')
```

```python
# Now, open the shapefile using pysal's FileIO
shps = ps.open(fpath , 'r')
```

```python
# We can read a polygon...
polygon = shps.next()
```

```python
shp_link = ps.examples.get_path('Chicago77.shp')
shp = ps.open(shp_link)
some = [bool(rdm.getrandbits(1)) for i in ps.open(fpath)]

fig = figure()

base = maps.map_poly_shp(shp)
base.set_facecolor('none')
base.set_linewidth(0.75)
base.set_edgecolor('0.8')
some = maps.map_poly_shp(shp, which=some)
some.set_alpha(0.5)
some.set_linewidth(0.)
cents = np.array([poly.centroid for poly in ps.open(fpath)])
pts = scatter(cents[:, 0], cents[:, 1])
pts.set_color('red')

ax = maps.setup_ax([base, some, pts], [shp.bbox, shp.bbox, shp.bbox])
fig.add_axes(ax)
show()

print 'Reading from ', shp_link
```
