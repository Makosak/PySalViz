## Moran's I Analysis

### Open data and the weight files.

```python
SIDS = ps.open(ps.examples.get_path("sids2.dbf"))
w = ps.open(ps.examples.get_path("sids2.gal")).read()
```

### Assign variable names from columns in data array

```python
BIR79 = np.array(SIDS.by_col('BIR79'))
SID79 = np.array(SIDS.by_col('SID79'))
SIDR79 = np.array(SIDS.by_col("SIDR79"))
```

### Calculate Moran's I

```python
mi = ps.esda.moran.Moran_Rate(SID79, BIR79, w, two_tailed=False)

# Moran Summary Report

print 'Observed Value: mi.i = {} '.format("%6.4f" % mi.I)
print 'Expected Value: mi.EI = {} '.format("%6.4f" % mi.EI)
print 'p-value: mi.p.norm = {} '.format("%6.4f" % mi.p_norm)
```

```python
np.random.seed(10)
mir = ps.esda.moran.Moran_Rate(SID79, BIR79, w, permutations = 9999)
```

```python
print mir.p_sim
print mir.EI_sim
print mir.z_sim
print mir.p_z_sim
```

## Plot Moran I as a regular scatter plot, or spatial lag versus our variable of interest.

### Determine Moran Plot quadrants for Data, and transform W for analysis

```python
# Z-Score standardisation
yt   = (SID79 - SID79.mean())/SID79.std()
ylt  = (w_SID79 - w_SID79.mean())/w_SID79.std()
```

```python
quad = np.zeros(yt.shape)
quad[np.bitwise_and(ylt > 0, yt > 0)]=1 # HH
quad[np.bitwise_and(ylt > 0, yt < 0)]=2 # LH
quad[np.bitwise_and(ylt < 0, yt < 0)]=3 # LL
quad[np.bitwise_and(ylt < 0, yt > 0)]=4 # HL
```

```python
w.transform = 'r'
```

### Calculate the spatial lag for the SIDS79 variable

```python
w_SID79= ps.lag_spatial(w, SID79)
#ols = ps.spreg.OLS(w_SID79[:, None], SID79[:, None])
```

```python
# Z-Score standardisation
y_std   = (SID79 - SID79.mean())/SID79.std()
yl_std  = (w_SID79 - w_SID79.mean())/w_SID79.std()
```

```python
# Check that we're returning an array of values
y_std
```

```python
## Operationialize a Moran Plot -- Very Basic:
## Input arguments: variable and weight
## maps.plot_moran(var, w, **kwargs)
## **kwargs: title='title', xlabel='xlabel', ylabel='ylabel', figsize=()

var = SID79
w = w
slag = ps.lag_spatial(w, var)

y_std   = (var - var.mean())/var.std()
yl_std  = (slag - slag.mean())/slag.std()

fig1 = plt.figure(figsize=(5,5))

plt.scatter(y_std, yl_std, s=60, color='k', alpha=.6)

plt.axvline(0, alpha=0.5)
plt.axhline(0, alpha=0.5)

plt.show()

## Add option for title and x & y labels in function
```

```python
fig2 = plt.figure(figsize=(8,8))

plt.scatter(y_std, yl_std, s=60, color='k', alpha=.6)

# Add title and labels
plt.suptitle('Moran Plot: SIDS 1979', fontsize=30)
plt.xlabel('SIDS79', fontsize=20)
plt.ylabel('SIDS79 Spatial Lag', fontsize=20)

# Add vertical and horizontal lines
plt.axvline(0, alpha=0.5)
plt.axhline(0, alpha=0.5)


plt.show()
```
